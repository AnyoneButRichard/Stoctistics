```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
ticker = yf.Ticker("MSFT")
df = ticker.history()
indices = [time.strftime("%m/%d/%y (%H:%M:%S)") for time in df.index]
```

```python
fig, ax = plt.subplots()
ax.plot(indices, df['High'])
fig.autofmt_xdate()
#plt.ylabel('some numbers')
plt.show()
```
