```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
fig_size = plt.rcParams.get('figure.figsize')
fig_size[0] = 20
fig_size[1] = 16
```

```python
ticker = yf.Ticker("AAPL")
df = ticker.history()
indices = [time for time in df.index]
```

```python
fig, ax = plt.subplots()
ax.plot(indices, df['High'])
fig.autofmt_xdate()
#plt.ylabel('some numbers')
plt.show()
```
